```python
#read trition output files
pieces0=[]
pieces1=[]
pieces2=[]

for y in np.nditer([power_density]):
    for x in np.nditer([burn_days]):
        frame0=pd.read_csv("C:/Users/tharr_000/Google Drive/uranium_break_out_risk_project/BSCM/bscm_TRITON_inputs/bscmTRITON_%1.2f_%d.00000000000000000.plt" % (y,x),header=None,index_col=0, delim_whitespace=True,skiprows=6).transpose().tail(1)
        frame1=pd.read_csv("C:/Users/tharr_000/Google Drive/uranium_break_out_risk_project/BSCM/bscm_TRITON_inputs/bscmTRITON_%1.2f_%d.00000000000000001.plt" % (y,x),header=None,index_col=0, delim_whitespace=True,skiprows=6).transpose().tail(1)
        frame2=pd.read_csv("C:/Users/tharr_000/Google Drive/uranium_break_out_risk_project/BSCM/bscm_TRITON_inputs/bscmTRITON_%1.2f_%d.00000000000000002.plt" % (y,x),header=None,index_col=0, delim_whitespace=True,skiprows=6).transpose().tail(1)            
        frame2['power_den'] = y
        frame2['burn_days'] = x
        pieces0.append(frame0)
        pieces1.append(frame1)
        pieces2.append(frame2)
        
pu0=pd.concat(pieces0,ignore_index=True)
pu1=pd.concat(pieces1,ignore_index=True)
pu2=pd.concat(pieces2,ignore_index=True)
tup0=pu0.shape
tup1=pu1.shape
tup2=pu2.shape
offset0=tup0[1]-1
offset1=tup1[1]-1
offset2=tup2[1]-3
offset3=tup2[1]-2
pudata=pu0.ix[:,:offset0].join(pu1.ix[:,:offset1]).join(pu2.ix[:,:offset2]).join(pu2.ix[:,offset3:])
#pudata=pudata.set_index(['power_den', 'burn_days'])
#pudata.pivot_table(values='pu239', index=['burn_days'], columns=['power_den'], aggfunc=sum).sort().plot()
#legend(loc=9, bbox_to_anchor=(1.5,1.2))
pudata['sum_pu_g']=pudata.ix[:,"pu236":"pu246"].sum(axis=1)
pudata['bscm_multipler'] = 8000/pudata.sum_pu_g
purange = np.arange(236,247,1)
for y in np.nditer([purange]):
    pudata['pu%d_bscm_g' % y] = pudata["pu%d" % y]*pudata['bscm_multipler']
for y in np.nditer([purange]):
    pudata['pu%d_20per_bscm_g' % y] = pudata["pu%d" % y]*pudata['bscm_multipler']*0.2
pudata['puheat_bscm_watts'] = pudata["pu"]*pudata['bscm_multipler']
pudata['puheat_wattsperkg'] = pudata["puheat_bscm_watts"]/8
pudata['puheat_20per_bscm_watts'] = pudata["pu"]*pudata['bscm_multipler']*0.2
pudata['quality'] = pudata["pu239"]/pudata['sum_pu_g']
pudata=pudata.set_index(['power_den','burn_days'])
```

```python
#read orgin output files
pieces0=[]
pu0=[]
for y in power_density:
    for x in burn_days:
        bscm_20perneu = 'C:/Users/tharr_000/Google Drive/uranium_break_out_risk_project/BSCM/bscm_20per_inputs/bscm20per_%1.2f_%d.00000000000000004.plt' % (y,x)
        bscm_20perpho = 'C:/Users/tharr_000/Google Drive/uranium_break_out_risk_project/BSCM/bscm_20per_inputs/bscm20per_%1.2f_%d.00000000000000001.plt' % (y,x)
        readerinneu_fluence=pd.read_csv(bscm_20perneu, header=None, skiprows=6, nrows=80, delim_whitespace=True)
        readerinpho_fluence=pd.read_csv(bscm_20perpho, header=None, skiprows=6, nrows=46, delim_whitespace=True)
        neu=readerinneu_fluence.reset_index().iloc[::2, 2:].reset_index().iloc[::, 1:].join(neucoeff).product(axis=1).sum(axis=0)*2*10**(-4)
        pho=readerinpho_fluence.reset_index().iloc[::2, 2:].reset_index().iloc[::, 1:].join(phocoeff).product(axis=1).sum(axis=0)*2*10**(-4)
        dose=(neu+pho)*3600*0.14/(10**(10))
        neusource=readerinneu_fluence.iloc[::2, 1:].join(readerinneu_binwidth).product(axis=1).sum(axis=0)*2*10**(-4)/2
        pieces0.append([y, x, dose, neusource]) 
pu0.extend(pieces0)
pu0=pd.DataFrame(pu0,columns = ['power_den','burn_days','dose','neusource'])
pu0=pu0.set_index(['power_den','burn_days'])
pudata=pudata.join(pu0)
pudata=pudata.reset_index()
pudata['runs_per_year']=pudata['burn_days'] / 365
pudata['FOM1']=1-log10(8/800+pudata['puheat_wattsperkg']*8/4500+8/50*(pudata['dose']/500)**(1/log10(2)))
pudata['FOM2']=1-log10(8/800+pudata['puheat_wattsperkg']*8/4500+pudata['neusource']*8/(6.8*10**6)+8/50*(pudata['dose']/500)**(1/log10(2)))
pudata['Pu_g/MTU_produced_per_Year'] = pudata['sum_pu_g'] *  pudata['runs_per_year']
pudata['MTU_needed_produce_8000gPu_in1year']= 8000 / pudata['Pu_g/MTU_produced_per_Year']
pudata['Pu_legal']=pudata['MTU_needed_produce_8000gPu_in1year']*Pu_legal
pudata['Pu_Japanese_Seawater']=pudata['MTU_needed_produce_8000gPu_in1year']*Pu_Japanese_Seawater
pudata['Pu_ORNL_Seawater']=pudata['MTU_needed_produce_8000gPu_in1year']*Pu_ORNL_Seawater
pudata['Pu_Coal_Ash_Acid_Leach']=pudata['MTU_needed_produce_8000gPu_in1year']*Pu_Coal_Ash_Acid_Leach
```

```python
#Create plots
dose=pudata.pivot('power_den', 'burn_days', 'dose')
X=dose.columns.values
Y=dose.index.values
Z=dose.values
x,y=np.meshgrid(X, Y)
CS = plt.contourf(x, y, Z, 15, cmap=cm.PuBu_r, linewidth=.5, locator=ticker.AutoLocator())
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel('Irradiation Time in Days', fontsize=14)
plt.ylabel('Power Density in MW/MTU', fontsize=14)
plt.title('Dose', fontsize=20)
cbar = plt.colorbar(CS)
cbar.ax.set_ylabel('Dose rad / hr')
plt.figure()

hS=plt.contourf(pudata.pivot('power_den', 'burn_days', 'MTU_needed_produce_8000gPu_in1year'), locator=ticker.LogLocator(), cmap=cm.PuBu_r)
plt.clabel(hS, inline=1, fontsize=10)
plt.xlabel('Irradiation Time in Days', fontsize=14)
plt.ylabel('Power Density in MW/MTU', fontsize=14)
plt.title('Natural U needed to produce 1 Sq of Pu a year', fontsize=20)
hbar = plt.colorbar(hS)
hbar.ax.set_ylabel('MTU')
plt.figure()

gS=plt.contourf(pudata.pivot('power_den', 'burn_days', 'FOM1'), locator=ticker.AutoLocator(), cmap=cm.PuBu_r)
plt.clabel(gS, inline=1, fontsize=10)
plt.xlabel('Irradiation Time in Days', fontsize=14)
plt.ylabel('Power Density in MW/MTU', fontsize=14)
plt.title('FoM 1', fontsize=20)
gbar = plt.colorbar(gS)
gbar.ax.set_ylabel('FOM 1')
plt.figure()

kS=plt.contourf(pudata.pivot('power_den', 'burn_days', 'FOM2'), locator=ticker.AutoLocator(), cmap=cm.PuBu_r)
plt.clabel(kS, inline=1, fontsize=10)
plt.xlabel('Irradiation Time in Days', fontsize=14)
plt.ylabel('Power Density in MW/MTU', fontsize=14)
plt.title('FoM 2', fontsize=20)
kbar = plt.colorbar(kS)
kbar.ax.set_ylabel('FOM 2')
plt.figure()

nS=plt.contourf(pudata.pivot('power_den', 'burn_days', 'Pu_Coal_Ash_Acid_Leach'), locator=ticker.LogLocator(), cmap=cm.PuBu_r)
plt.clabel(nS, inline=1, fontsize=10)
plt.xlabel('Irradiation Time in Days', fontsize=14)
plt.ylabel('Power Density in MW/MTU', fontsize=14)
plt.title('Price for 1 sq from coal ash', fontsize=20)
nbar = plt.colorbar(nS)
nbar.ax.set_ylabel('Price in dollars')
plt.figure()
```
